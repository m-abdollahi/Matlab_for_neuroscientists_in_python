```python
#import libraries
import scipy.io
mat = scipy.io.loadmat('Chap17_Data.mat')
import numpy as np
import matplotlib.pyplot as plt
```

```python
#store Variables and data from Matlab to Python
instruction=mat["instruction"]                     #instruction  times
go =mat["go"]                                      #Go cue times
unit=mat["unit"]["times"][0]                       #unit times
dir = mat["direction"]                             #Dircetions
```

```python
#creating a dictionry for Differernt directions and storing their index
directions_group_inx = {}
for i in [1,2,3,4,5,6,7,8]:
    directions_group_inx["direction{0}".format(i)] = []
    
for i in range(len(instruction)):
    if dir[i] == 1:
        directions_group_inx["direction1"].append(i)
    elif dir[i]==2:
            directions_group_inx["direction2"].append(i)
    elif dir[i]==3:
            directions_group_inx["direction3"].append(i)
    elif dir[i]==4:
            directions_group_inx["direction4"].append(i)
    elif dir[i]==5:
            directions_group_inx["direction5"].append(i)
    elif dir[i]==6:
            directions_group_inx["direction6"].append(i)
    elif dir[i]==7:
            directions_group_inx["direction7"].append(i)
    elif dir[i]==8:
            directions_group_inx["direction8"].append(i)
```

```python
#creatin a dictionarfor storing the times of spikes
#directions_group_time=np.array([[],[],[],[],[],[],[],[]],dtype=object)
directions_group_time={}
for i in [1,2,3,4,5,6,7,8]:
    directions_group_time["direction{0}".format(i)] = {}
    

   
#mega for loop !!    
for i in [1,2,3,4,5,6,7,8]:
    for ii in range(len(instruction[directions_group_inx["direction{0}".format(i)]])):
        b=directions_group_time["direction{0}".format(i)]["trial{0}".format(ii)] = []
        for iii in range(2):
            for iiii in unit[iii]:
                a = instruction[directions_group_inx["direction{0}".format(i)]]
                #b = directions_group_time["direction{0}".format(i)]
                if iiii>= a[ii] and iiii<= a[ii]+1:
                    b.append(iiii) 
    
           
```

```python
#normalize
normed=directions_group_time
d1t1 = np.squeeze(normed["direction1"]["trial1"])
d1t1 = (d1t1-d1t1[0])/(d1t1[-1]-d1t1[0])
d2t1 = np.squeeze(normed["direction2"]["trial1"])
d2t1 = (d2t1-d2t1[0])/(d2t1[-1]-d2t1[0])
d3t1 = np.squeeze(normed["direction3"]["trial1"])
d3t1 = (d3t1-d3t1[0])/(d3t1[-1]-d3t1[0])
d4t1 = np.squeeze(normed["direction4"]["trial1"])
d4t1= (d4t1-d4t1[0])/(d4t1[-1]-d4t1[0])
d5t1= np.squeeze(normed["direction5"]["trial1"])
d5t1 = (d5t1-d5t1[0])/(d5t1[-1]-d5t1[0])
d6t1= np.squeeze(normed["direction6"]["trial1"])
d6t1 = (d6t1-d6t1[0])/(d6t1[-1]-d6t1[0])
d7t1= np.squeeze(normed["direction7"]["trial1"])
d7t1 = (d7t1-d7t1[0])/(d7t1[-1]-d7t1[0])
d8t1= np.squeeze(normed["direction8"]["trial1"])
d8t1 = (d8t1-d8t1[0])/(d8t1[-1]-d8t1[0])
```

```python
#RasterPLot
plt.subplot(3,3,1)
plt.eventplot(d1t1)
plt.title("Direction1")
plt.xlabel("time")
plt.ylabel("units spike")

#RasterPLot
plt.subplot(3,3,2)
plt.eventplot(d2t1)
plt.title("Direction2")
plt.xlabel("time")
plt.ylabel("units spike")


#RasterPLot
plt.subplot(3,3,3)
plt.eventplot(d3t1)
plt.title("Direction3")
plt.xlabel("time")
plt.ylabel("units spike")


#RasterPLot
plt.subplot(3,3,4)
plt.eventplot(d4t1)
plt.title("Direction1")
plt.xlabel("time")
plt.ylabel("units spike")

#RasterPLot
plt.subplot(3,3,6)
plt.eventplot(d5t1)
plt.title("Direction5")
plt.xlabel("time")
plt.ylabel("units spike")

#RasterPLot
plt.subplot(3,3,7)
plt.eventplot(d6t1)
plt.title("Direction6")
plt.xlabel("time")
plt.ylabel("units spike")

#RasterPLot
plt.subplot(3,3,8)
plt.eventplot(d7t1)
plt.title("Direction7")
plt.xlabel("time")
plt.ylabel("units spike")

#RasterPLot
plt.subplot(3,3,9)
plt.eventplot(d8t1)
plt.title("Direction8")
plt.xlabel("time")
plt.ylabel("units spike")


```

```python
d1fr=len(directions_group_time_peth["direction1"])/len(directions_group_inx_peth["direction1"])
d2fr=len(directions_group_time_peth["direction2"])/len(directions_group_inx_peth["direction2"])
d3fr=len(directions_group_time_peth["direction3"])/len(directions_group_inx_peth["direction3"])
d4fr=len(directions_group_time_peth["direction4"])/len(directions_group_inx_peth["direction4"])
d5fr=len(directions_group_time_peth["direction5"])/len(directions_group_inx_peth["direction5"])
d6fr=len(directions_group_time_peth["direction6"])/len(directions_group_inx_peth["direction6"])
d7fr=len(directions_group_time_peth["direction7"])/len(directions_group_inx_peth["direction7"])
d8fr=len(directions_group_time_peth["direction8"])/len(directions_group_inx_peth["direction8"])
```

```python
## ploting tuning curve
plt.subplots()
plt.plot([1,2,3,4,5,6,7,8],[d1fr,d2fr,d3fr,d4fr,d5fr,d6fr,d7fr,d8fr],label="Actuall data")
plt.title("Tuning Curve")
plt.xlabel("direction")
plt.ylabel("spike count firing rate")

#cosine estimator / model
from scipy.optimize import curve_fit
def data(x,g,c):
    final = g *np.cos(x)+c
    return final

## fitting crue with scipy function see https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html for more details!
x= [1,2,3,4,5,6,7,8]
z = [d1fr,d2fr,d3fr,d4fr,d5fr,d6fr,d7fr,d8fr]
popt,_= curve_fit(data,x,z)
g,c = popt 
print(f"the offset is equl to = {c}",f"the eff is equl to = {g}")
plt.plot(x,3*np.cos([0,1,2,3,4,5,6,7])+c,label="simulated model")
plt.legend()
```
